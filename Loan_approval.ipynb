```python
df= pd.get_dummies(df,columns=[' education',' self_employed'],drop_first=True)

df =df.astype(int)
```

```python
df.head()
```

```python
X= df.drop(columns=[' loan_status','loan_id'],axis=1)
Y = df[' loan_status']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.20, random_state=42)
```

```python
scaler= StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled= scaler.transform(X_test)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
Log = LogisticRegression()

Log.fit(X_train_scaled,Y_train)
```

```python
Y_pred = Log.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
acc=accuracy_score(Y_test,Y_pred)
acc
```

```python
confusion_matrix(Y_test,Y_pred)
```

```python
print(classification_report(Y_test,Y_pred))
```

# Stacking

```python
from sklearn.ensemble import StackingClassifier
```

```python
base_learners= [
    ('model_dt', DecisionTreeClassifier(random_state=42)),
    ('model_svm',SVC(probability=True, kernel='rbf',random_state=42)),
    ('Log',LogisticRegression(max_iter=1000))
]
```

```python
meta_learner = KNeighborsClassifier()
```

```python
stacking_clf= StackingClassifier(
    estimators=base_learners,
    final_estimator=meta_learner,
    cv=5
)
```

```python
stacking_clf.fit(X_train,Y_train)
```

```python
Y_pred_st= stacking_clf.predict(X_test)
```

```python
acc_st= accuracy_score(Y_pred_st,Y_test)
acc_st
```

```python
confusion_matrix(Y_pred_st,Y_test)
```

```python
print(classification_report(Y_pred_st,Y_test))
```

# Comparison of Accuracy Scores

```python
import plotly.express as px
import pandas as pd

models = ['Logistic Regression', 'KNN', 'SVM',
          'Naive Bayes', 'Decision Tree',
          'Random Forest', 'Stacking',
          'Boosting (AdaBoost)']

scores = [acc, acc_knn, acc_svm, acc_nb,
          acc_dt, acc_rf, acc_st, acc_ada]

results = pd.DataFrame({
    'Model': models,
    'Accuracy': scores
})

fig = px.bar(
    results,
    x='Model',
    y='Accuracy',
    text='Accuracy',
    title='Comparison of Classification Models'
)

fig.update_traces(texttemplate='%{text:.2f}', textposition='outside')

fig.update_layout(
    xaxis_title='Models',
    yaxis_title='Accuracy (%)',
    xaxis_tickangle=-45
)

fig.show()
```
